```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#unique players into a variable
Total_Players=purchase_data["SN"].unique()
#Make a dictionary
Total_Players={"Total Players":[len(Total_Players)]}
#load the dictionary into a dataframe for reporing
Total_Players_df=pd.DataFrame(Total_Players,columns=["Total Players"])
Total_Players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculate each column individualy in separate variables
UniqueItems=len(purchase_data["Item ID"].unique())
Price=purchase_data["Price"].mean()
Purchases=purchase_data["Purchase ID"].count()
Rev=purchase_data["Price"].sum()
#create a dataframe with the values of the variables
summaryDF=pd.DataFrame({"Number of Unique Items":[UniqueItems], "Average Price": [Price], "Number of Purchases":[Purchases], "Total Revenue":[Rev]})
#format the columns
summaryDF["Average Price"]=summaryDF["Average Price"].map("${:.2f}".format)
summaryDF["Total Revenue"]=summaryDF["Total Revenue"].map("${:.2f}".format)
summaryDF
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#find distinct total number of players
Total_Players=purchase_data["SN"].nunique()
#group the data by gender
GroupByDf=purchase_data.groupby("Gender")
#get unique counts of each row
TotalCount=GroupByDf.nunique()
#Add columns to the dataframe to represent what we need to the final report
TotalCount["Percentage of Players"]= (TotalCount["SN"]/Total_Players)*100
TotalCount["Total Count"]=TotalCount["SN"]
#Select only the relevant columns from the dataframe
SummaryTable=TotalCount[["Total Count","Percentage of Players"]]
#sort
SummaryTable = SummaryTable.sort_values(["Total Count"], ascending=False)
#format
SummaryTable["Percentage of Players"]=SummaryTable["Percentage of Players"].map("{:.2f}".format)
#display
SummaryTable
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#group by gender and agregate the appropiate columns
AvgDF=purchase_data.groupby("Gender").agg({"Purchase ID":['nunique'], "Price":["mean",'sum'],"SN":"nunique"})
#add columns for additional calcs
AvgDF["Avg Total Purchase per Person"]=AvgDF["Price"]["sum"]/AvgDF["SN"]["nunique"]
AvgDF["Purchase Count"]=AvgDF["Purchase ID"]["nunique"]
AvgDF["Average Purchase Price"]=AvgDF["Price"]["mean"]
AvgDF["Total Purchase Value"]=AvgDF["Price"]["sum"]

#format
AvgDF["Avg Total Purchase per Person"]=AvgDF["Avg Total Purchase per Person"].map("${:.2f}".format)
AvgDF["Average Purchase Price"]=AvgDF["Average Purchase Price"].map("${:.2f}".format)
AvgDF["Total Purchase Value"]=AvgDF["Total Purchase Value"].map("${:.2f}".format)

#Select only the relevant columns
FinalTable=AvgDF[["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]]
FinalTable
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#unique players into a variable
Total_Players=len(purchase_data["SN"].unique())

#list and lables for the bins in 2 lists
Values=[0,9,14,19,24,29,34,39,122]
Names=["<10","10-14", "15-19","20-24","25-29","30-34","35-39","40+"]
#apply bins to data on Age
purchase_data["Age Bracket"] = pd.cut(purchase_data["Age"], Values, labels=Names)
# group by age and count distinct players
AgeDF=purchase_data.groupby("Age Bracket").agg({"SN":["nunique"]})
#Create a % column 
AgeDF["Percentage of Players"]=(AgeDF["SN"]["nunique"]/Total_Players)*100
AgeDF["Total Count"]=AgeDF["SN"]["nunique"]
#format
AgeDF["Percentage of Players"]=AgeDF["Percentage of Players"].map("{:.2f}".format)

#display
AgeDF[["Total Count","Percentage of Players"]]
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

#list and lables for the bins in 2 lists
Values=[0,9,14,19,24,29,34,39,122]
Names=["<10","10-14", "15-19","20-24","25-29","30-34","35-39","40+"]

#apply bins to data on Age
purchase_data["Age Bracket"] = pd.cut(purchase_data["Age"], Values, labels=Names)

#group by age and count distinct players
AgeDF=purchase_data.groupby("Age Bracket").agg({"SN":["nunique"], "Price":["mean","sum"], "Purchase ID":['nunique']})

#Create  column 
AgeDF["Purchase Count"] =AgeDF["Purchase ID"]["nunique"]
AgeDF["Average Purchase Price"] =AgeDF["Price"]["mean"]
AgeDF["Total Purchase Value"] =AgeDF["Price"]["sum"]
AgeDF["Avg Total Purchase per Person"]=(AgeDF["Price"]["sum"])/(AgeDF["SN"]["nunique"])

#format
AgeDF["Average Purchase Price"]=AgeDF["Average Purchase Price"].map("${:.2f}".format)
AgeDF["Total Purchase Value"]=AgeDF["Total Purchase Value"].map("${:.2f}".format)
AgeDF["Avg Total Purchase per Person"]=AgeDF["Avg Total Purchase per Person"].map("${:.2f}".format)

#display
AgeDF[["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]]
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

#group distinct players
AgeDF=purchase_data.groupby("SN").agg({ "Price":["mean","sum"], "Purchase ID":['nunique']})

#Create  column 
AgeDF["Purchase Count"] =AgeDF["Purchase ID"]["nunique"]
AgeDF["Average Purchase Price"] =AgeDF["Price"]["mean"]
AgeDF["Total Purchase Value"] =AgeDF["Price"]["sum"]

#format
AgeDF["Average Purchase Price"]=AgeDF["Average Purchase Price"].map("${:.2f}".format)
AgeDF["Total Purchase Value"]=AgeDF["Total Purchase Value"].map("${:.2f}".format)

#display
AgeDF[["Purchase Count","Average Purchase Price","Total Purchase Value"]].sort_values(["Total Purchase Value"] , ascending=False).head(5)
```

## Most Popular Items

```python
#purchase_data.columns
#Group by item id and item name, then aggregate
PopDF=purchase_data.groupby(["Item ID","Item Name"]).agg({"Item ID":["count"],"Price":["mean","sum"]})
#Create New columns form the new aggregations and formats
PopDF["Purchase Count"]=PopDF["Item ID"]["count"]
PopDF["Item Price"]=PopDF["Price"]["mean"].map("${:.2f}".format)
PopDF["Total Purchase Value"]=PopDF["Price"]["sum"].map("${:.2f}".format)
#Same column unformated for sorting
PopDF["Total Sort"]=PopDF["Price"]["sum"]
#display top 5

PopDF[["Purchase Count","Item Price","Total Purchase Value"]].sort_values(["Purchase Count"], ascending=False).head(5)
```
